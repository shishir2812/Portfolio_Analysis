### Creating Dataframe

#### Raw Data Table

```python
import pandas as pd
import numpy as np
from datetime import datetime
import requests

filename = 'Activities_for_31Dec2019_to_22Sep2021.xlsx'
USD_to_CAD_rate = 1.26
#TODO: automate^ 
def print_current_df(dataframe):
    with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
        display(dataframe)
```

# Set Up

#### Desired DataFrame

1. Convert Settlement Date to DateTime
2. Drop Transaction Date, Account #, Activity Type, Account Type


```python
raw_data_table = pd.read_excel(filename, index_col=False)
df = raw_data_table.drop(columns=['Transaction Date', 'Account #', 'Activity Type', 'Account Type', 'Description']).sort_values('Settlement Date')
# remove Norbert's Gambits transactions
df = df.loc[(df['Symbol'] != 'DLR') & (df['Symbol'] != 'DLR.TO') & (df['Symbol'] != 'H038778')]
df.rename(columns = {'Settlement Date': 'Date'}, inplace = True)


#if your index is not in datetime type yet, use the following before you try to extract year, month and day.
#https://stackoverflow.com/questions/55776571/how-to-split-a-date-column-into-separate-day-month-year-column-in-pandas
df['Date'] = pd.to_datetime(df.Date)
# df.set_index('Date', inplace= True)
print_current_df(df)
```

### Total Money Deposited and Total Commission



```python
total_commission = df['Commission'].sum()
total_deposited = df.groupby('Action')['Net Amount'].sum()['CON']
total_commission, total_deposited
```

```python
print(f'total deposited: {total_deposited}')
print(f'total commision paid: {-total_commission}')
print(f'total holding:')
df.dropna().sort_values(['Symbol','Date'])
```

## All in CAD

```python
cad_df = df
cad_df['Net Amount'] = np.where(cad_df['Currency'] == 'USD',cad_df['Net Amount']*USD_to_CAD_rate, cad_df['Net Amount']).round(2)
cad_df.drop(columns = ['Currency'], inplace = True)
cad_df
```

```python
# cad_df = cad_df[cad_df['Action'] != 'CON']
# cad_df = cad_df[cad_df['Action'] != 'FCH']
transactions_df = cad_df.loc[(cad_df['Action'] == 'Buy') | (cad_df['Action'] == 'Sell')]
transactions_df.sort_values('Symbol')
```

```python
transactions_df.groupby('Symbol').agg({'Net Amount':'sum'}).sort_values('Net Amount')
```
